```python
import pandas as pd 
import numpy as np
```

```python
summary = pd.read_csv('Cora_summary.csv')
result_df = summary.copy()
import matplotlib.pyplot as plt

x = result_df['Epoch'].values
y = result_df['Loss'].values

plt.plot(x, y)
plt.xlabel('Epoch')
plt.ylabel('Cora Loss')
plt.grid(axis = 'y')
plt.show()
```
